```python
df = pd.read_csv("/Users/Vidya/Projects/datasets/neruda.txt", sep = '/', header = None)
print(df.info())
```

```python
df = df.transpose()
```

```python
df[0:20]
```

```python
df.rename(columns = {0: 'poems'}, inplace= True)
```

```python
stopwords = set(STOPWORDS)
```

```python
df
```

```python
text = " ".join(review for review in df.poems)
```

```python
wordcloud = WordCloud(stopwords= stopwords,width = 2000, height = 1000,max_font_size=150, max_words=100, background_color="white").generate(text)
```

```python
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.savefig("/Users/Vidya/Projects/datasets/neruda_wordcloud.png", format="png")
plt.show()
```
